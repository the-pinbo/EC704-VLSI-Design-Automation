```python
import graphviz
```

```python
g = graphviz.Graph('G', filename='g_c_n.gv')
g.attr(label='m2 m3 v m1')
g
```

```python
with g.subgraph(name='cluster1') as c1:
    g.attr(label='m2 m3')
    with g.subgraph(name='cluster2') as c2:
        with g.subgraph(name='cluster3') as c3:
            c3.node('m2')
        with g.subgraph(name='cluster4') as c4:
            c4.node('m2')
        c2.node('m3')
    c1.node('m1')
    
g
```

```python
with g.subgraph(name='cluster1') as c:
    c.attr(fillcolor='blue:cyan', label='acluster', fontcolor='white',
           style='filled', gradientangle='270')
    c.attr('node', shape='box', fillcolor='red:yellow',
           style='filled', gradientangle='90')
    c.node('anode',width='2.0',height='2.0')
    c.node('cathode',width='4.0',height='1.0')
    c.node('test',width='1.0',height='4.0')


g
```
