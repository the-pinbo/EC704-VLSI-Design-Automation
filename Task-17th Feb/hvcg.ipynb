```python
#importing libraries
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#routing vectors
l1=[1,2,3,2,5,1,4,5]
l2=[7,3,7,4,3,6,4,6]
```

```python

#horizontal constraint graph
rows=max(l1+l2)
cols=max(l1+l2)
adjh = [[0 for i in range(cols)] for j in range(rows)]
#print(adjh)
for i in range(len(l2)):
    val=l2[i]
    for j in l2[i+1:]:
        if j==val:
            idx=[p for p, n in enumerate(l2) if n == j][1]
            #print(idx)
            for k in range(i,(idx+1)):
                if val!=l1[k]:
                   adjh[val-1][l1[k]-1]=1 
                   adjh[l1[k]-1][val-1]=1
                #print(adjh)
adjh
```

```python
#vertical constraint graph
rows=max(l1+l2)
cols=max(l1+l2)
adjv = [[0 for i in range(cols)] for j in range(rows)]
adjv
for i in range(len(l1)):
    if l1[i]!=l2[i]:
        adjv[l1[i]-1][l2[i]-1]=1
#mylabels=[row[0] for row in adjv]
adjv
#mylabels
```

```python
#plotting graphs
#hcg
Gh=nx.Graph()
for i in range(rows):
    for j in range(cols):
        if adjh[i][j]==1:
            Gh.add_edge(i+1,j+1)
nx.draw(Gh,with_labels=True)
plt.show()

#vcg
Gv=nx.DiGraph()
for i in range(rows):
    for j in range(cols):
        if adjv[i][j]==1:
            Gv.add_edge(i+1,j+1)
nx.draw(Gv,with_labels=True)
plt.show()
```
